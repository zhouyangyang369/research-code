```python
import pandas as pd
from  collections import defaultdict
import requests

city_path1 = pd.read_excel('/Users/yangyang-z/documents/python/apple_music/城市URL和sql语句.xlsx','Sheet10',usecols=[7],names=None,header=None)
city_path2 = pd.read_excel('/Users/yangyang-z/documents/python/apple_music/城市URL和sql语句.xlsx','Sheet10',usecols=[10],names=None,header=None)
city = city_path1[7].values.tolist()
city2 = city_path2[10].values.tolist()

for j in range(102,len(city)):
    city_song = pd.read_excel(city[j])
    song1 = city_song['SONG'].values.tolist()
    song = list(set(song1)) #去重
    df = pd.DataFrame(index = song, columns = ['danceability','energy','loundiness','valence','tempo'])
    null_id = []
    songs_id = {}
    for i in range(len(song)):
        params['q'] = song[i]
        print(song[i])
        response = requests.get(url_search, headers=headers, params=params)
        print(response)
        track = response.json()
        print(track)
        tracks = track['tracks']
        items_list = tracks['items']
        if len(items_list) != 0:
            items_dict = items_list[0]
            id1 = items_dict['id']
            songs_id[song[i]] = id1
        else:
            null_id.append(song[i])
        
    songs_audio_features = {}
    for key1 in songs_id.keys():
        song_id1 = songs_id[key1]
        url = ['https://api.spotify.com/v1/audio-features/']
        url[0] = url[0] + song_id1
        response = requests.get(url[0], headers=headers)
        songs_audio_features[key1] = response.json()

    for key in songs_audio_features.keys():
        features = songs_audio_features[key]
        #print(features)
        if 'error' in features.keys():
            danceability = 0
            energy = 0
            loudness = 0
            valence = 0
            tempo = 0
        else:
            danceability = features['danceability']
            energy = features['energy']
            loudness = features['loudness']
            valence = features['valence']
            tempo = features['tempo']
        df.loc[key] = [danceability,energy,loudness,valence,tempo]
    print(city[j])
    df.to_excel(city2[j])
    df.drop(df.index,inplace=True)
```
